```python
file = open('../data/spotify.html',mode='r')
html = file.read()
file.close()
```

```python
import re

urls = [ m.group() for m in re.finditer('https?://\S+(?=")', html) ]
hex_colors = [ m.group() for m in re.finditer('#[a-fA-F0-9]{6,6}', html)]
artists = [ m.group() for m in re.finditer('(?<=<span>by ).*(?=</span>)', html) ]
```

```python
import json

urlsJson = json.dumps(urls)
file = open('urls.json', 'w')
file.write(urlsJson)
file.close()

hex_colorsJson = json.dumps(hex_colors)
file = open('hex.json', 'w')
file.write(hex_colorsJson)
file.close()

artistsJson = json.dumps(artists)
file = open('artist.json', 'w')
file.write(artistsJson)
file.close()
```
